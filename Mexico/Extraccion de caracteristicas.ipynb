```python
data2 = data.copy()
#Calculamos el numero total de las poblaciones "n"
n = 8269

# Convertimos la columna a tipo numérico eliminando el símbolo '%'
data2['host_response_rate'] = data2['host_response_rate'].str.rstrip('%').astype(float)

#Obtenemos el limite superior y el limite inferior de la columna objetivo
Max_host_response_rate = data2['host_response_rate'].max()
Min_host_response_rate = data2['host_response_rate'].min()

#Calculamos el rango R
R_host_response_rate = Max_host_response_rate - Min_host_response_rate
#R_host_response_rate

#Calculamos el numero de Intervalos de Clase "ni", aplicando la regla de Sturges
ni_host_response_rate = 1 + 3.32 * np.log10(n)
#ni_host_response_rate

# Limitamos el valor de ni entre 5 y 12
ni_host_response_rate = max(5, min(12, round(ni_host_response_rate)))

#Calculamos el ancho del intervalo "i"
i_host_response_rate = R_host_response_rate/ni_host_response_rate
#i_host_response_rate

#Categorizacion de variables
intervalos = np.linspace(Min_host_response_rate, Max_host_response_rate, ni_host_response_rate + 1)
#intervalos

#Creamos las categorias un poco mas descriptivas
categorias = [f"Tasa de {intervalos[i]:.2f}-{intervalos[i+1]:.2f}" for i in range(len(intervalos) - 1)]

#Finalmente creamos las categorias en la columna numerica
data2['host_response_rate'] = pd.cut(x = data2['host_response_rate'], bins = intervalos, labels = categorias)
#data2['host_response_rate']

#Creamos el grafico de barras para las frecuencias
host_response_rate = freq_tbl(data2['host_response_rate'])
host_response_rate2 = host_response_rate.drop(['percentage', 'cumulative_perc'], axis = 1)
Filtro = host_response_rate2[host_response_rate2['frequency'] > 1]
Filtro_index = Filtro.set_index('host_response_rate')
Filtro_index.plot(kind = 'bar', width = 0.8, figsize = (10,4), alpha = 0.5, color = 'red')
plt.title('Tasa de respuesta del host')
plt.xlabel('Tasa')
plt.ylabel('Frecuencia')
```
